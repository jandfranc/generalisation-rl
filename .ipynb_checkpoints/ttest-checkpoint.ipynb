```python
word_1 = ['old','new']
compare_pre_post_df = pd.DataFrame()
compare_pre_post_df['Agent'] = [1,2,3,4,5]

for first in word_1:
    list_t = []
    list_p = []
    for num in range(1,6):
        preword = f'{first}_pre_{num}'
        postword = f'{first}_post_{num}'
        
        
        t_num, p_num = stats.ttest_ind(df[preword],df[postword], equal_var=False)
        list_t.append(t_num)
        list_p.append(p_num)
    
    compare_pre_post_df[first + '_t'] =list_t 
    compare_pre_post_df[first + '_p'] =list_p 

compare_pre_post_df
        
        
        
```

```python
import numpy as np

def get_p_star(value):
    if value < 0.0001:
        return '***'
    if value < 0.01:
        return '***'
    if value < 0.05:
        return '***'
    return ''

pre_old_df = np.empty((5,5))
pre_new_df = np.empty((5,5))
post_old_df = np.empty((5,5))
post_new_df = np.empty((5,5))
pre_old_df_p = np.empty((5,5))
pre_new_df_p = np.empty((5,5))
post_old_df_p = np.empty((5,5))
post_new_df_p = np.empty((5,5))

for num_1 in range(1,6):        
    for num_2 in range(1,6):
        t_num, p_num = stats.ttest_ind(df[f'old_pre_{num_1}'],df[f'old_pre_{num_2}'], equal_var=False)
        pre_old_df[num_1-1,num_2-1] = t_num
        pre_old_df_p[num_1-1,num_2-1] = p_num
        
        t_num, p_num = stats.ttest_ind(df[f'new_pre_{num_1}'],df[f'new_pre_{num_2}'], equal_var=False)
        pre_new_df[num_1-1,num_2-1] = t_num
        pre_new_df_p[num_1-1,num_2-1] = p_num
        
        t_num, p_num = stats.ttest_ind(df[f'old_post_{num_1}'],df[f'old_post_{num_2}'], equal_var=False)
        post_old_df[num_1-1,num_2-1] = t_num
        post_old_df_p[num_1-1,num_2-1] = p_num
        
        t_num, p_num = stats.ttest_ind(df[f'new_post_{num_1}'],df[f'new_post_{num_2}'], equal_var=False)
        post_new_df[num_1-1,num_2-1] = t_num
        post_new_df_p[num_1-1,num_2-1] = p_num
        
                

```

```python
def create_df(nparray):
    return pd.DataFrame(nparray, index=[1,2,3,4,5], columns=[1,2,3,4,5])

def create_stars(t_array, p_array):
    text_array = np.empty((5,5), dtype='U100')
    for num_1 in range(5):
        for num_2 in range(5):
            stars = get_p_star(p_array[num_1,num_2])
            t = str(round(t_array[num_1,num_2],4))
            text_array[num_1,num_2] = t + stars
    return text_array
```

```python
pre_old_df_star = create_df(create_stars(pre_old_df, pre_old_df_p))
pre_new_df_star = create_df(create_stars(pre_new_df,pre_new_df_p))
post_old_df_star = create_df(create_stars(post_old_df, post_old_df_p))
post_new_df_star = create_df(create_stars(post_new_df, post_new_df_p))
```
